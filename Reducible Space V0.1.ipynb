```python
def apply_exclusions(df, exclusion_register):
    """
    Remove rows from a pandas DataFrame based on exclusion criteria
    specified in a dictionary.

    Parameters:
        df (pandas.DataFrame): The DataFrame to filter.
        exclusion_register (dict): A dictionary containing the columns to
        filter and the exclusion criteria for each column.

    Returns:
        pandas.DataFrame: The filtered DataFrame.
    """
    filtered_df = df.copy()  # make a copy of the input DataFrame to avoid modifying it
    
    for col, exclusion in exclusion_register.items():
        if col in filtered_df.columns:
            
            if isinstance(exclusion[0], str):
                # checks whether the exclusion data is a string
                filtered_df = filtered_df[~filtered_df[col].isin(exclusion)]
                
            elif isinstance(exclusion[0], (int, float)) and len(exclusion) == 2:
                # this checks if the data is numerical and the right length
                filtered_df = filtered_df[(filtered_df[col] >= exclusion[0]) &
                                          (filtered_df[col] <= exclusion[1])]
                
    return filtered_df                                                                        
```

```python
def split_index(FAS_df):
    new_index = [(FAS_1 if (layer := FAS_df.iat[n, 0]) == 1 else FAS_1, index)
                 for n, index in enumerate(FAS_df.index) if (FAS_1 := index) or layer == 2]

    FAS_df.set_index(pd.MultiIndex.from_tuples(new_index, names=['FAS 1', 'FAS 2']), inplace=True)
    FAS_df = FAS_df[FAS_df['FAS Layer'] > 1]
    FAS_df = FAS_df.drop(columns=["FAS Layer"])
    
    return FAS_df


def split_index1(FAS_df):
    new_index = []
    FAS_1 = None

    for n in range(len(FAS_df['FAS Layer'])):
        layer = FAS_df.iat[n,0]
        index = FAS_df.index[n]

        if layer == 1:
            FAS_1 = index       
            new_index.append((FAS_1, index))
        elif layer == 2:
            new_index.append((FAS_1, index))

    # set new index and drop rows where FAS Layer == 1
    FAS_df.set_index(pd.MultiIndex.from_tuples(new_index, names=['FAS 1', 'FAS 2']), inplace=True)
    
    FAS_df = FAS_df[FAS_df['FAS Layer'] > 1]
    # Not needed, but kept for reference: FAS_df.drop(FAS_df[FAS_df['FAS Layer'] != 2].index, inplace=True)
    FAS_df.drop(columns=["FAS Layer"], inplace=True)
    
    return FAS_df
```

```python
# This is the function that applies core/support categories to the rows in the Pri/Sec dataframes

def core_support(FAS_df, FAS_2_categories):
    
    core_support_column = []
    
    FAS_1_values = FAS_df.index.get_level_values(0).unique()
    FAS_2_values = FAS_df.index.get_level_values(1)
    
    for FAS_2 in FAS_2_values:
        if FAS_2_categories.at[FAS_2, "Space classification"] == 'Core':
            core_support_column.append('Core')
            
        elif FAS_2_categories.at[FAS_2, "Space classification"] == 'Support':
            core_support_column.append('Support')
            
        else:
            core_support_column.append('N/A')
    
    FAS_df.insert(0, "Space classification", core_support_column)
    return FAS_df
```

```python
# This is the function that calculates enrolment band. Inputs are enrolments and upper enrolment band (to provide flexibility incase this changes)

def enrolment_category(enrolment, max_val):
    # Generate a range of values at intervals of 25 using numpy
    bins = np.arange(0, max_val+25, 25)

    if enrolment in bins:
        category = enrolment
        # Use numpy digitize to categorize the enrollment
    else:
        category_idx = np.digitize(enrolment, bins)
        # Return the highest value in the category
        category = bins[category_idx] if category_idx > 0 else None

    return category
```

```python
# This is a function that calculates enrolment_categories: this will need to be added into the class

def enrolmentment_categories(campus, school_data, FAS_1_categories, FAS_2_categories, Sec_FAS, Pri_FAS):
    
    # this gets the school based on the campus input, which is used to index the school_data dataframe
    school = school_data.index.get_level_values(0)[school_data.index.get_level_values(1) == campus].item()
    
    primary_FAS_max_enrolments = (len(list(Pri_FAS.columns)) - 2) * 25 # this finds the number of the FAS categories (i.e. maximum number of primary students in the FAS)
    
    secondary_FAS_max_enrolments = (len(list(Sec_FAS.columns)) - 2) * 25 # this finds the number of the FAS categories (i.e. maximum number of secondary students in the FAS)
    
    primary_enrolments = school_data.loc[(school,campus), "Primary Enrolments"].item() # this retrieves the number of primary enrolments from the school_data for the given campus
    
    secondary_enrolments = school_data.loc[(school,campus), "Secondary Enrolments"].item() # this retrieves the number of secondary enrolments from the school_data for the given campus
    
    primary_enrolment_category = enrolment_category(primary_enrolments, primary_FAS_max_enrolments) # this finds the enrolment band for the primary FAS
    
    secondary_enrolment_category = enrolment_category(secondary_enrolments, secondary_FAS_max_enrolments) # this finds the enrolment band for the secondary FAS
    
    enrolment_categories = (primary_enrolment_category, secondary_enrolment_category)
        
    return enrolment_categories

enrolment_categories = enrolmentment_categories("Burwood College : Pri/Sec", school_data, FAS_1_categories, FAS_2_categories, Sec_FAS, Pri_FAS)

```

```python
# This returns the entitled space for both primary and secondary FAS 2 categories as two dataframes in a tuple)

def FAS2_entitled_space(enrolment_categories, Pri_FAS, Sec_FAS):
    
    primary_enrolments_category = enrolment_categories[0]
    
    secondary_enrolments_category = enrolment_categories[1]
    
    primary_entitled = pd.DataFrame(data = 0, index = pd.MultiIndex.from_tuples(list(Pri_FAS.index)), columns = ["Entitled Space"])
    
    secondary_entitled = pd.DataFrame(data = 0, index = pd.MultiIndex.from_tuples(list(Sec_FAS.index)), columns = ["Entitled Space"])
    
    for i in list(primary_entitled.index):
        primary_entitled.at[i, "Entitled Space"] += Pri_FAS.at[i,str(primary_enrolments_category)]
    
    for i in list(secondary_entitled.index):
        secondary_entitled.at[i, "Entitled Space"] += Sec_FAS.at[i,str(secondary_enrolments_category)]
        
    primary_entitled = core_support(primary_entitled, FAS_2_categories)
    
    secondary_entitled = core_support(secondary_entitled, FAS_2_categories)
    
    return (primary_entitled, secondary_entitled)
    

FAS_entitlement = FAS2_entitled_space(enrolment_categories, Pri_FAS, Sec_FAS)
```
